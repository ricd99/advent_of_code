```python
with open("input.txt", "r") as f:
    ranges = f.read().split(",")
ranges
```

```python
# part 1
def sum_fake_codes(input):
    sum = 0
    for r in ranges:
        for i in range(int(r.split("-")[0]), int(r.split("-")[1]) + 1):
            i = str(i)
            mid = len(i) // 2
            
            front = i[:mid]
            back =  i[mid:]
            
            if front == back:
                sum += int(i)
    return(sum)

sum_fake_codes(ranges)
```
