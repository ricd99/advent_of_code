```python
with open("input.txt", "r") as f:
    grid = [list(line) for line in f]
    print(grid)
    
```

```python


def accessible_rolls(grid):
    total = 0
    keep_going = True
    while keep_going:
        curr_count = 0
        for i in range(len(grid)):   
            for j in range(len(grid[i])):
                if grid[i][j] == "@":
                    count = 0
                    try:
                        if i > 0 and j > 0 and grid[i-1][j-1] == "@":
                            count+=1
                    except IndexError:
                        pass
                    try:
                        if i > 0 and grid[i-1][j] == "@":
                            count+=1
                    except IndexError:
                        pass
                    try:
                        if i > 0 and grid[i-1][j+1] == "@":
                            count+=1
                    except IndexError:
                        pass
                    try:
                        if j > 0 and grid[i][j-1] == "@":
                            count+=1
                    except IndexError:
                        pass
                    try:
                        if grid[i][j+1] == "@":
                            count+=1
                    except IndexError:
                        pass
                    try:
                        if j > 0 and grid[i+1][j-1] == "@":
                            count+=1
                    except IndexError:
                        pass
                    try:
                        if grid[i+1][j] == "@":
                            count+=1
                    except IndexError:
                        pass
                    try:
                        if grid[i+1][j+1] == "@":
                            count+=1
                    except IndexError:
                        pass
                    if count < 4:
                        total +=1
                        curr_count += 1
                        grid[i][j] = "."
                        
    return total

accessible_rolls(grid)
```
